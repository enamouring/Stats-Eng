```python
import numpy as np #importing libraries
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10,20) #linear array with 20 points between -10 to 10
```

```python
x
```

```python
m = 50 #slope
b = 17 #y-intercept
```

```python
y = m * x + b #slope-intercept form
```

```python
plt.plot(x,y,'.') #plot the line with the corresponding inputs
```

```python
r = 100*np.random.randn(20,) #random 'noisy' data (y-values)
```

```python
r
```

```python
plt.plot(r,'.') #plotting r
```

```python
y2 = m * x + b + r #adding r to our original equation (y2)
```

```python
plt.plot(x,y2,'.') #simulation of real data
```

```python
m_guess = 1 #initial guess (slope)
b_guess = 1 #inital guess (y-intercept)

y_guess = m_guess * x + b_guess #generate a line with the corresponding guesses

plt.plot(x,y_guess,'r.');
plt.plot(x,y2,'g.'); #plots the guess line and y2
```

```python
for m_guess in np.linspace(-100,100,100): #iterate testing different slope values to find the best fitting line

        y_guess = m_guess * x #to generate the line using current m guess

        plt.plot(x,y_guess,'r.'); #plots the guess line and y2
        plt.plot(x,y2,'g.');
        plt.show()


        loss = np.sum((y_guess - y)**2) #computes the loss between guess line and y2 by doing (guess-y2)^2

        print(loss,'M = ',m_guess) #print the loss and current guess line
```

```python
best_loss = 100000000000 #initialize best loss, m, and b (y-intercept)
best_m = 0
best_b = 0

for m_guess in np.linspace(-100,100,100): #guessing m and b combinations to create the best fit line
    for b_guess in np.linspace(-100,100,100):

        y_guess = m_guess * x + b_guess #equation to find our new best fit line, iterated for all m_guess and b_guess

        loss = np.sum((y_guess - y)**2) #calulcate the loss between y_guess and y (our original line)

        if loss < best_loss: #if loss is less than best_loss update the best_loss, best_m, and best_b to be the values from their guesses and loss

            best_loss = loss
            best_m = m_guess
            best_b = b_guess
```

```python
y_guess = best_m * x + best_b #shows the best fit line

plt.plot(x,y_guess,'r.'); #plots the best fit line and our original line
plt.plot(x,y2,'g.');
plt.show()

print(best_loss,best_m,best_b) #prints the best_loss, best_m, and best_b values
```

```python
import numpy as np #import libraries
import matplotlib.pyplot as plt
```

```python
X = np.array([19,65,66,80,97,113,175,270,310,350,350,650,680]) #x points
Y = np.array([1150,3800,5000,4800,5400,6700,10300,15400,19500,23200,21600,39400,41000]) #y points
```

```python
plt.plot(X,Y,'g.') #plots the points with their correpsonding x and y values (in green)
```

```python
m_guess = 0 #base guess for slope
b_guess = 0 #base guess for y-intercept
```

$y=mx+b$

```python
m_guess * X + b_guess #calulates the slope guess values times each x value plus the b guess (initial y values)
```

```python
y_guess = m_guess * X + b_guess
```

```python
y_guess - Y #calculates the difference between the guess and the given values
```

```python
np.sum((y_guess - Y)**2) #calculates the sum of all the differences between the guess and the given values squared
```

```python
loss = np.sum((y_guess - Y)**2) #assigns this sum to be the loss
```

```python
best_loss_so_far = 1000000000000 #initalize best values
m_best = 0
b_best = 0
for m_guess in np.linspace(-100,100,100): #iterates through possible slope and y-intercept values
    for b_guess in np.linspace(-100,100,100):

        y_guess = m_guess * X + b_guess #computes the y-value corresponding to each guess line

        loss = np.sum((y_guess - Y)**2) #determines the loss

        if loss < best_loss_so_far: #if the loss is less than the best loss, replace all best values with the new values obtained
            m_best = m_guess
            b_best = b_guess
            best_loss_so_far = loss

print("Found best fit:")
print(m_best,b_best,loss) #displays the best fit line
```

```python
m_best #states the best slope
```

```python
H = m_best #assigns the slope to H for Hubble's constant
```

```python
1/H #calculates the multiplicative inverse of H
```

If the universe has been expanding since its beginning at a constant speed, the universe's age would simply be 1/Ho.
Find the inverse of your value of Ho.
Multiply the inverse by $3.09 \times 10^{19}$ km/Mpc to cancel the distance units.
Since you now have the age of the Universe in seconds, divide this number by the number of seconds in a year: $3.16 \times 10^{7}$ sec/yr.

$3.09 \times 10^{19}$ km/Mpc

$3.16 \times 10^{7}$ sec/year

```python
(1/H) * (3.09*10**19) #estimated age of the universe in seconds
```

```python
(1/H) * (3.09*10**19)  / (3.16*10**7) #estimated age of the universe in years
```

```python
 (1/H) * (3.09*10**19)  / (3.16*10**7) / (1000000000) #estimated age of the universe in billions of years
```

```python
import pandas as pd #import libraries
import numpy as np
import matplotlib.pyplot as plt
def get_database(url,sheet_name): #gets a database from a url and sheet name
    url_head = "https://docs.google.com/spreadsheets/d/"
    url_foot = "/gviz/tq?tqx=out:csv&sheet="
    url_body = url.split('/')[5]
    # sheet_name = "1"
    url_csv = url_head + url_body + url_foot + sheet_name
    df = pd.read_csv(url_csv, sep=',', skiprows=0)
    return df
```
